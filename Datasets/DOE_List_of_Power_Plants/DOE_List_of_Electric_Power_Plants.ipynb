Import libraries pandas and numpy

```python
import pandas as pd
import numpy as np
```

save excel files as dataframes and format column names to be in a single row

```python
luzon_df=pd.concat(pd.read_excel('electric_power_plants_luzon_december_2020-converted.xlsx',sheet_name=[0,1,2,3,4,5],skiprows=1, header=[0,1]), ignore_index=True, sort=False)
visayas_df=pd.concat(pd.read_excel('electric_power_plants_visayas_december_2020-converted.xlsx',sheet_name=None,skiprows=0,header=[1,2]), ignore_index=True, sort=False)
mindanao_df=pd.concat(pd.read_excel('electric_power_plants_mindanao_december_2020-03_may-2021-converted.xlsx',sheet_name=[0,1],skiprows=None,header=[0,1]), ignore_index=True, sort=False)
offgrd_df=pd.concat(pd.read_excel('electric_power_plants_lvm_offgrid_december_2020-converted.xlsx',sheet_name=None,skiprows=1,header=[0,1]), ignore_index=True, sort=False)
```

```python
luzon_df.columns=['_'.join(col) for col in luzon_df.columns.to_flat_index()]
visayas_df.columns=['_'.join(col) for col in visayas_df.columns.to_flat_index()]
mindanao_df.columns=['_'.join(col) for col in mindanao_df.columns.to_flat_index()]
offgrd_df.columns=['_'.join(col) for col in offgrd_df.columns.to_flat_index()]
offgrd_df.rename(columns={'LOCATION_MUN/PROV':'LOCATION_MUNICIPALITY/ PROVINCE'}, inplace=True)
offgrd_df
```

combine dataframes created for luzon, visayas, mindanao, and off-grid power plants

```python
all_df=luzon_df[["POWER PLANT_FACILITY NAME","POWER PLANT_SUBTYPE","CAPACITY, MW_DEPENDABLE","LOCATION_MUNICIPALITY/ PROVINCE"]].append(visayas_df[["POWER PLANT_FACILITY NAME","POWER PLANT_SUBTYPE","CAPACITY, MW_DEPENDABLE","LOCATION_MUNICIPALITY/ PROVINCE"]]).append(mindanao_df[["POWER PLANT_FACILITY NAME","POWER PLANT_SUBTYPE","CAPACITY, MW_DEPENDABLE","LOCATION_MUNICIPALITY/ PROVINCE"]]).append(offgrd_df[["POWER PLANT_FACILITY NAME","POWER PLANT_SUBTYPE","CAPACITY, MW_DEPENDABLE","LOCATION_MUNICIPALITY/ PROVINCE"]])
all_df
```

trim data to get only the location of solar power plants

```python
solar=all_df.loc[all_df["POWER PLANT_SUBTYPE"].str.contains("solar", na=False, case=False)].sort_values('CAPACITY, MW_DEPENDABLE',ascending=False)
solar_loc=solar[['LOCATION_MUNICIPALITY/ PROVINCE']]
solar_loc.rename(columns={'LOCATION_MUNICIPALITY/ PROVINCE':'LOCATION'},inplace=True)
solar_loc
```
