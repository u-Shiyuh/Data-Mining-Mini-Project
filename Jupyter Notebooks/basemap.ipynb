```python
import os
os.environ['PROJ_LIB'] = r'C:\Users\Shiyuh\.conda\pkgs\proj4-5.2.0-hc56fc5f_1006\Library\share'
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import ScalarMappable
from mpl_toolkits.basemap import Basemap
from matplotlib.animation import FuncAnimation
```

```python
df = pd.read_csv('formaps_2.csv')
df
```

```python
### Create a figure and draw a basemap

fig, ax = plt.subplots(figsize=(5,4), dpi=500)

m = Basemap(resolution='h',
            projection='lcc',
            lat_0=11.9, lon_0=122.5,
            llcrnrlon=117, llcrnrlat= 5, urcrnrlon=127, urcrnrlat=19)

m.scatter(df['Longitude'][df['Cluster_kmeans']==0].tolist(),df['Latitude'][df['Cluster_kmeans']==0].tolist(),latlon=True,s=20, label="cluster 1", color='red')
m.scatter(df['Longitude'][df['Cluster_kmeans']==1].tolist(),df['Latitude'][df['Cluster_kmeans']==1].tolist(),latlon=True,s=20, label="cluster 2", color='green')
m.scatter(df['Longitude'][df['Cluster_kmeans']==2].tolist(),df['Latitude'][df['Cluster_kmeans']==2].tolist(),latlon=True,s=20, label="cluster 3", color='blue')
plt.title("K means clustering", color='white')
plt.legend(loc="upper left", fontsize=4.5)
m.drawcoastlines()
m.drawmapboundary(zorder=0)
#m.fillcontinents(color='#ffffff',zorder=1)
m.drawcountries(linewidth=1.5)
m.drawstates()

```
